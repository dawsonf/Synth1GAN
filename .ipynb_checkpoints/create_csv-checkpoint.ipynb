```python
def read_synth1_file(filepath):
    try:
        f = open(filepath)
        file = f.read()
        f.close()
        file=file.split("\n")
        header = file[0:3]
        data = file[3:-1]
        synth_data = {}
        synth_data["pack"] = filepath.split("/")[-2]
        synth_data["name"] = header[0]
        synth_data["color"] = header[1].split("=")[1]
        synth_data["ver"] = header[2].split("=")[1]
        data_values = {x[0]:x[1] for x in [x.split(",") for x in data]}
        synth_data.update(data_values)
        
    
        return synth_data
    except:
        #print("Error Decoding")
        return None
    
```

```python
def read_synth1_folder(folderpath):
    res = glob.glob(folderpath+"*.sy1")

    synth_presets = []
    for i in range(len(res)):
        r = read_synth1_file(res[i])
        if r:
            synth_presets.append(r)
    
    return synth_presets
```

```python
def read_all_folders(mega_folderpath):
    res = glob.glob(mega_folderpath+"/*/")
    master_list = []
    
    for i in res:
        #print(i)
        master_list.extend(read_synth1_folder(i))
    #ls = read_synth1_folder(res[0])
    
    return master_list
    
    #print(res)
```
