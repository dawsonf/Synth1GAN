```python
import pandas
import os
```

```python
df = pandas.read_csv("presets_ver106.csv")
df = df.drop(columns=["pack","name","color","ver"])
```

```python
batch_num = 18
```

```python
to_convert = pandas.read_csv("./out2/"+str(batch_num)+".csv")
```

```python
def un_normalize(source_df, target, stdev=False):
    target.columns = list(source_df.columns)
    if stdev:
        u_normalized = (target*(source_df.std())+source_df.mean())
    else:
        u_normalized = ( ((target+1)*(source_df.max() - source_df.min()) )/2)+source_df.min()
    return u_normalized.round(0).astype(int)
```

```python
result = un_normalize(df, to_convert)
```

```python
result
```

```python
def write_row_to_file(num,data):
    d_version = data.iloc[num].to_dict()
    to_write = []
    to_write.append("SynthGAN-"+str(num))
    to_write.append("color=red")
    to_write.append("ver=106")
    for i in d_version.keys():
        to_write.append(i+','+str(d_version[i]))
        
    with open("batch_"+str(batch_num)+"/"+str(num).zfill(3)+".sy1", 'w') as fo:
       for d in to_write:
         fo.write(str(d) + '\n')
```

```python
os.makedirs("batch_"+str(batch_num),exist_ok=True)

for i in range(len(result)):
    write_row_to_file(i,result)
```
